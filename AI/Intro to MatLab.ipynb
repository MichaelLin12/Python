```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y,'r-')
plt.show()
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#OOP
fig = plt.figure()
 # left,bottom,width,height
axes = fig.add_axes([0.1,0.1,1,1])
axes.plot(x,y)
axes.set_xlabel('Y')
axes.set_ylabel('X')
axes.set_title('Title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

ax1.plot(x,y)
ax1.set_title('larger plot')
ax2.plot(y,x)
ax2.set_title('smaller plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for curr in axes:
    curr.plot(x,y)
    curr.set_title("Title")
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout
```

```python
fig = plt.figure(figsize=(3,2),dpi=100)

axe = fig.add_axes([0,0,1,1])
axe.plot(x,x**2,label='x^2')
axe.plot(x,x**3,label='x^3')
axe.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=2,alpha=0.5,linestyle='--',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')#RGB HEX CODE
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
